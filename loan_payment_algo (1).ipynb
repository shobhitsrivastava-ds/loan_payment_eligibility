```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Loan payments data.csv")
```

```python
data.Gender=data.Gender.map({"male":1,"female":0})
```

```python
a1=pd.get_dummies(data["education"])
```

```python
a1.head()
```

```python
data=pd.concat([data,a1],axis="columns")
```

```python
data.drop(["education","college"],axis="columns",inplace=True)
```

```python
data.loan_status.values
```

```python
data.loan_status=data.loan_status.map({"PAIDOFF":0,"COLLECTION":1,"COLLECTION_PAIDOFF":2})
```

```python
data.head()
```

```python
data.past_due_days.values
```

```python
data.past_due_days=data.past_due_days.fillna(0)
```

```python
data.drop("Loan_ID",axis="columns",inplace=True)
```

```python
data.head()
```

```python
data.Principal.values
```

```python
X=data.drop(["loan_status","paid_off_time","effective_date","due_date"],axis="columns")
```

Data parsing :

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.87)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
type(X_test)
```

Model Selection :

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=10)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
score=metrics.accuracy_score(y_test,y_pred)
```

```python
score
```

```python
metrics.confusion_matrix(y_test,y_pred)
```
